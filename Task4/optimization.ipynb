## Optimization Model using Linear Programming


A manufacturing company produces two products: A and B.  
The goal is to maximize profit using limited machine and labor resources.


Let x = units of Product A  
Let y = units of Product B  

Maximize: 40x + 30y  

Subject to:  
2x + y ≤ 40  
x + 2y ≤ 60  
x ≥ 0 , y ≥ 0


```python
from pulp import *

# Create model
model = LpProblem("Profit_Maximization", LpMaximize)

# Decision variables
x = LpVariable('Product_A', lowBound=0)
y = LpVariable('Product_B', lowBound=0)

# Objective function
model += 40 * x + 30 * y

# Constraints
model += 2 * x + y <= 40
model += x + 2 * y <= 60

# Solve
model.solve()

# Output
print("Status:", LpStatus[model.status])
print("Product A units:", x.varValue)
print("Product B units:", y.varValue)
print("Maximum Profit = ₹", value(model.objective))
```
